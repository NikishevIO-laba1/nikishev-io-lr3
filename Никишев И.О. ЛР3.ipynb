```python
import lr3
```

```python
input_vars = {
    'вес': 0.7,
    'ширина': 13.5,
    'длина': 27,
}

f = lr3.Fuzzy(rules = lr3.yaml_load('продукционные правила.yaml'), 
              objects = lr3.yaml_load('объекты.yaml'), 
              variables = lr3.yaml_load('лингвистические переменные.yaml'), 
              input_vars = input_vars)
```

```python
f.plot_graphs()
```

```python
f.fuzzificate()
```

```python
f.aggregate()
```

```python
f.activate()
```

```python
f.accumulate()
```
